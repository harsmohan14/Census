```python
dfM1 = df.ix[:,['AAGE', 'ACLSWKR', 'ADTIND', 'ADTOCC', 'AHGA', 'AHSCOL',
       'AMARITL', 'AMJIND', 'AMJOCC', 'ARACE', 'AREORGN', 'ASEX', 'AUNMEM',
       'AUNTYPE', 'AWKSTAT', 'FILESTAT','GRINREG', 'GRINST', 'HHDFMX', 'HHDREL', 
        'MIGMTR1', 'MIGMTR3','MIGMTR4', 'MIGSAME', 'MIGSUN', 'NOEMP', 'PARENT', 'PEFNTVTY',
       'PEMNTVTY', 'PENATVTY', 'PRCITSHP', 'SEOTR', 'VETQVA', 'VETYN',
       'WKSWORK', 'YEAR', 'Outcome']]
```

```python
dfM1.shape
```

```python
dfM1.columns
```

```python
#Doing cross tabulation for categorical features
```

```python
dfM1_Cross1 = pd.get_dummies(dfM1,columns=['ACLSWKR', 'ADTIND', 'ADTOCC', 'AHGA', 'AHSCOL', 'AMARITL',
       'AMJIND', 'AMJOCC', 'ARACE', 'AREORGN', 'ASEX', 'AUNMEM', 'AUNTYPE',
       'AWKSTAT', 'FILESTAT', 'GRINREG', 'GRINST', 'HHDFMX', 'HHDREL',
       'MIGMTR1', 'MIGMTR3', 'MIGMTR4', 'MIGSAME', 'MIGSUN', 'NOEMP', 'PARENT',
       'PEFNTVTY', 'PEMNTVTY', 'PENATVTY', 'PRCITSHP', 'SEOTR', 'VETQVA',
       'VETYN', 'WKSWORK', 'YEAR'], drop_first=True)
dfM1_Cross1.shape
```

```python
dfM1_Cross = dfM1_Cross1.iloc[:,dfM1_Cross1.columns != 'HHDFMX_ Grandchild <18 ever marr not in subfamily']
dfM1_Cross.shape
```

```python
def convertOutcomeToInts(x):
    if x == ' - 50000.':
        x = 0
    else:
        x = 1
    return x
    
```

```python
dfFinalM1 = dfM1_Cross
dfFinalM1.shape
```

```python
dfFinalM1.Outcome = dfM1_Cross.Outcome.apply(convertOutcomeToInts)
dfFinalM1.Outcome.unique()
```

```python
dfFinalM1_Train, dfFinalM1_Test = train_test_split(dfFinalM1,test_size=0.2,random_state=10)
```

```python
dfFinalM1_Train.shape
```

```python
dfFinalM1_Test.shape
```

```python
dfFinalM1_Train_X = dfFinalM1_Train.iloc[:,dfFinalM1_Train.columns != 'Outcome']
dfFinalM1_Train_Y = dfFinalM1_Train.Outcome
```

```python
dfFinalM1_Test_X = dfFinalM1_Test.iloc[:,dfFinalM1_Test.columns != 'Outcome']
dfFinalM1_Test_Y = dfFinalM1_Test.Outcome
```

```python
dfFinalM1_Train_X.shape
```

```python
dfFinalM1_Train_Y.unique()
```

```python
#First Using Randomforest to see variables importance
from sklearn.ensemble import RandomForestClassifier
```

```python
rfClassifier = RandomForestClassifier(n_estimators=100,n_jobs=1)
```

```python
rfClassifier.fit(dfFinalM1_Train_X,dfFinalM1_Train_Y)
```

```python
important_features = pd.Series(data=rfClassifier.feature_importances_,index=dfFinalM1_Train_X.columns)
important_features.sort_values(ascending=False,inplace=True)
```

```python
important_features
```

```python
predicted_rf_M1 = rfClassifier.predict(dfFinalM1_Test_X)
```

```python
from sklearn import metrics
```

```python
print (metrics.classification_report(predicted_rf_M1, dfFinalM1_Test_Y))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
results_M1 = cross_val_score(rfClassifier, dfFinalM1_Test_X, dfFinalM1_Test_Y, cv=20)
print(results_M1.mean())
```
